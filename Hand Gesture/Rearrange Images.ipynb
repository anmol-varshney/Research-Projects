```python
import cv2
import os
```

```python
os.curdir
```

```python
BASE_PATH = os.path.join(os.curdir, "MP_DATA")
print(BASE_PATH)
```

```python
for dir in os.listdir(BASE_PATH):
    CLASS_PATH = os.path.join(BASE_PATH, dir)
    i=0
    for vid in os.listdir(CLASS_PATH):
        VID_PATH = os.path.join(CLASS_PATH, vid)
        for frame in os.listdir(VID_PATH):
            ImagePath = os.path.join(VID_PATH, frame)
            image = cv2.imread(ImagePath)
            ImageName = "{}.png".format(i)
            print(ImageName)
            WRITE_PATH = os.path.join(CLASS_PATH, ImageName)
            cv2.imwrite(WRITE_PATH, image)
            i+=1
```
